```python
import numpy as np
import matplotlib.pyplot as plt

# 
initial_conditions = [100, 12]  # &
t_start = 0
t_end = 10
t = np.linspace(t_start, t_end, 100)  # 

#  Lotka-Volterra 
def lotka_volterra(t, y):
    X, Y = y
    # alphabeta
    alpha = 0.5
    beta = 0.8
    # 100%0
    delta_X = 1.0
    delta_Y = 0.0
    dXdt = alpha * X * delta_X * Y
    dYdt = -delta_Y * X * Y
    return [dXdt, dYdt]

# 
from scipy.integrate import solve_ivp
solution = solve_ivp(lotka_volterra, [t_start, t_end], initial_conditions, t_eval=t)

# &
X = solution.y[0]
Y = solution.y[1]

# 
plt.plot(t, X, label='Defender (X)')
plt.plot(t, Y, label='Threat & Suspicious (Y)')
plt.xlabel('Time')
plt.ylabel('Quantity')
plt.title('Lotka-Volterra Model')
plt.legend()
plt.grid(True)

# 
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp
import scienceplots

plt.style.use(['science','ieee','no-latex'])
def lotka_volterra(t, y, alpha, beta, gamma, delta):
    x, y = y
    dxdt = alpha*x - beta*x*y
    dydt = delta*x*y - gamma*y
    return [dxdt, dydt]

# 
alpha = 0.8  # 
beta = 0.2   # 
gamma = 0.4  # 
delta = 0.1  # 

# 
x0 = 100  # 
y0 = 50   # 
initial_conditions = [x0, y0]

# 
t_start = 0
t_end = 100
t_step = 0.01
t = np.arange(t_start, t_end, t_step)

# 
solution = solve_ivp(
    fun=lambda t, y: lotka_volterra(t, y, alpha, beta, gamma, delta),
    t_span=(t_start, t_end),
    y0=initial_conditions,
    t_eval=t
)

# 
x = solution.y[0]
y = solution.y[1]

# 
plt.plot(t, x, label='Threat Detection System')
plt.plot(t, y, label='Prey (Suspicious and Threat Nodes)')
plt.xlabel('Time')
plt.ylabel('Population')
plt.title('Lotka-Volterra Model')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import comb

# Given normal probabilities and malicious probabilities
normal_probs = [0.00014588403918048184, 0.00013838634724174436, 0.0001623401607664638, 0.00019107603701843566]
malicious_probs = [0.000000001505876625381683, 0.0000000042696182877256425, 0.000000004937666866926433, 0.000000001948889839144988]

# Define the probability functions
def P_pass(k, n, f, c, r):
    prob = 0
    for i in range(k, n + 1):
        prob += comb(n, i) * (f + (1 - f) * c / r) ** i * ((1 - f) * (1 - c)) ** (n - i)
    return prob

def P_fp(k, n, f):
    prob = 0
    for i in range(k):
        prob += comb(n, i) * f ** (n - i) * (1 - f) ** i
    return prob

# Define parameter values
k = 1
n = 4
f_values = np.linspace(0, 1, 100)
c = 0.5
r = 2

# Calculate probabilities
pass_probs = [P_pass(k, n, f, c, r) for f in f_values]
fp_probs = [P_fp(k, n, f) for f in f_values]

# Plot the results
plt.plot(f_values, pass_probs, label='P_pass')
plt.plot(f_values, fp_probs, label='P_fp')
plt.scatter(normal_probs, malicious_probs, color='red', label='Data Points')
plt.xlabel('f')
plt.ylabel('Probability')
plt.legend()
plt.title('Probability Functions')
plt.grid(True)
plt.show()
```
